```python
import urllib.request
import urllib.parse # 转译中文， python无法识别ascii码以外的编码
import string
```

```python
def get_method_params():
    url = "http://www.baidu.com/s?wd="    
    name = "美女"
    
    # 拼接字符串(汉字)
    temp_url = url + name
    print(temp_url)
    
    # 转译中文 解释器问题
    encode_url = urllib.parse.quote(temp_url, safe=string.printable)
    print(encode_url)
    
    # 使用代码发送网络请求
    response = urllib.request.urlopen(encode_url)
    print(response)
    
    # 读取内容
    data = response.read().decode()
    print(data)
    
    # 保存到本地
    with open("02_get_params.html","w",encoding="utf-8") as file:
        file.write(data)
    
```

```python
get_method_params()
```
