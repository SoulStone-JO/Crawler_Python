```python
import urllib.request
```

```python
url ="https://blog.csdn.net/rookie_is_me/article/details/81634048"
```

```python
# 添加代理
proxy = {
    # 免费的写法
#   "http":"http://182.107.176.243:9999"
    # 简写方法
    "http":"182.107.176.243:9999"
    # 付费的IP
}
```

```python
# 代理处理器
proxy_handler = urllib.request.ProxyHandler(proxy)
```

```python
# 创建自己的opener
opener = urllib.request.build_opener(proxy_handler)
```

```python
# 拿着代理IP发送请求
data = opener.open(url).read()
print(data.decode("utf-8"))
```
